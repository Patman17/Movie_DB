Creating dataframe of discover high revenue US movies 

```python
for i in range(0,500):
    page = i+1
    url = "https://api.themoviedb.org/3/discover/movie"
    querystring = {"api_key":{api_key},"language":"en-US","sort_by":"revenue.desc","page":{page}}

    headers = {
        'X-RateLimit': "4",
#         'User-Agent': "PostmanRuntime/7.16.3",
#         'Accept': "*/*",
#         'Cache-Control': "no-cache",
#         'Postman-Token': "31bdc574-98ca-499d-953b-be49e9ab5d4c,06b4b282-e37c-407b-999f-e483689c0fd3",
#         'Host': "api.themoviedb.org",
#         'Accept-Encoding': "gzip, deflate",
#         'Connection': "keep-alive",
#         'cache-control': "no-cache"
        }
    if i==0: 
        response = requests.request("GET", url, headers=headers, params=querystring)
        df = pd.DataFrame.from_dict(response.json()['results'])
    elif i >0:
        response2 = requests.request("GET", url, headers=headers, params=querystring)
        df2 =pd.DataFrame.from_dict(response2.json()['results'])
        df = df.append(df2)
    
    
```
